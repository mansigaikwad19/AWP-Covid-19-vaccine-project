```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('vacc.csv')
```

### lets see null values in the dataset

```python
df.isnull().sum()
```

#### filling the null values with zeroes and dropping the values to avoid the data giving us misleading information

```python
df.fillna(0, inplace = True)
df.drop(df.index[df['iso_code'] == 0], inplace = True)
```

```python
df.info()
```

## changing the datatype of date so we can work with the date as well

```python
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
df.info()
```

```python
df.describe()
```

# dropping these columns because we dont need them

```python
df.drop(['source_website', 'source_name','daily_vaccinations_per_million','daily_vaccinations_raw'], axis = 1)
```

# here we have detailed covid vaccination data of 223 countries

```python
df['country'].nunique()
```

### Now lets find top 10 countries who have vaccinated most num of people

```python
con_vac=df.groupby('country').max().sort_values('total_vaccinations', ascending=False)
con_vac = con_vac.iloc[:10]                 #iloc 10 for top 10 
con_vac
```

# Representing the information graphically

```python
plt.figure(figsize=(15, 6))
plt.bar(con_vac.index, con_vac.total_vaccinations)
plt.xticks(rotation = 90)
plt.ylabel('Total Vaccinations')
plt.xlabel('top 10 Countries')
plt.title("Top 10 countries who have vaccinated most number of people");
```

### Lets find out for vaccinations per 100 people

```python
con_vac100=df.groupby('country').max().sort_values('total_vaccinations_per_hundred', ascending=False)
con_vac100 = con_vac100.iloc[:10]                 #iloc 10 for top 10 
con_vac100
```

```python
plt.figure(figsize=(18, 6))
plt.bar(con_vac100.index, con_vac100.total_vaccinations_per_hundred)
plt.xticks(rotation = 90)
plt.ylabel('Total Vaccinations')
plt.xlabel('Top 10 Countries')
plt.title("top 10 countries with highest vaccinations per hundred people");
```

### lets try to understand the data in detail ...
### Lets start with
# India

```python
#saving the data of India in separate DataFrame
df_INDIA = df[df["iso_code"] == 'IND'].copy()
df_INDIA
```

#### Removing the null i.e 0's from the data

```python
df_INDIA
```

## vaccination in India till 2022

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=df_INDIA, x="date", y="total_vaccinations")
plt.title("Total vaccinations in INDIA")
plt.xticks(rotation=45);
```

## Daily vaccination rate in India till mid-jan

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=df_INDIA, x="date", y="daily_vaccinations")
plt.xticks(rotation=90)
plt.title("Daily vaccinations in INDIA");
```

# Graphically Represesnting how many people are remaning to be fully vaccinated

```python
y1=df_INDIA.loc[:,'total_vaccinations']
x1=df_INDIA.loc[:,"date"]
y2=df_INDIA.loc[:,'people_fully_vaccinated']
x2=df_INDIA.loc[:,"date"]
```

```python
plt.figure(figsize=(18,6))
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.legend(["Total Vaccinations", "People Fully Vaccinated"])
plt.title("Difference between Fully vaccinated people and total vaccinations in India");
```

# Now lets study the Data based on the Vaccines used

```python
# grouping by vaccines
vacc_con = df.groupby('vaccines').max().sort_values('total_vaccinations', ascending=False)
vacc_con=vacc_con.iloc[:10]
vacc_con
```

```python
#Reset indexx index from vaccines to countries as it will be easier to sort and plot based on countries
vacc_con=vacc_con.reset_index()
vacc_con
```

```python
plt.figure(figsize=(18,6))

sns.barplot(data = vacc_con, x='vaccines', y = 'total_vaccinations', hue = 'country', dodge=False)
plt.xticks(rotation=90);
plt.title("top 10 vaccines by country");
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data = vacc_con, x='vaccines', y = 'total_vaccinations_per_hundred', hue = 'country', dodge=False)
plt.xticks(rotation=90);
plt.title("top 10 vaccines by country");
```
